```python
from sklearn.utils import resample
zero = train[train['smoking']==0]
one = train[train['smoking']==1]
new = resample(train[train['smoking']==0], n_samples=70000, replace=True, random_state=3)
new['smoking'].value_counts()
```

```python
train = pd.concat([new, one])
train['smoking'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

ind = train['id']
X = train.drop(['id', 'smoking'], axis=1)
y = train['smoking']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_cols = X_train.select_dtypes(include='number').columns.tolist()
scaler.fit(X_train[num_cols])
X_train[num_cols] = scaler.transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
X
```

```python
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(X_train, y_train)
```

```python
roc_auc_score(y_test, clf.predict(X_test))
```
